## Distribution


### Determine how the data is distributed


The column `ConvertedComp` contains Salary converted to annual USD salaries using the exchange rate on 2019-02-01.

This assumes 12 working months and 50 working weeks.


Plot the distribution curve for the column `ConvertedComp`.


```python
df["ConvertedComp"].isna().sum()
df["ConvertedComp"].fillna(value=57745, inplace=True)  
```

```python
# your code goes here
df["ConvertedComp"].describe()
data = df["ConvertedComp"]
datasort = data.sort_values()
normfit = ss.norm.fit(data)

plt.figure(figsize=(8,6))
plt.plot(datasort, ss.norm.pdf(datasort, *normfit))
plt.title("ConvertedComp Normal Distribution")
plt.xlabel("Data Distribution")
plt.ylabel("Values")
plt.grid()
plt.show()
          
```

Plot the histogram for the column `ConvertedComp`.


```python
# your code goes here
plt.figure(figsize=(8,6))
plt.hist(data, density=True)
plt.title("Histogram Distribution of ConvertedComp")
plt.xlabel("ConvertedComp Data")
plt.ylabel("Values")
plt.show()
```

Find out the  median ConvertedComp of responders identified themselves only as a **Woman**?


```python
# your code goes here
women_cc = []

for Gender, ConvertedComp in zip(df["Gender"], df["ConvertedComp"]):
    if Gender == "Woman":
        women_cc.append(ConvertedComp)

print("Meadian ConvertedComp of responders identified as Woman: ", stats.median(women_cc))        
```

## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
plt.figure(figsize=(4,6))
plt.boxplot(df["ConvertedComp"])
plt.title("Boxplot of ConvertedComp")
plt.xlabel("Data of ConvertedComp")
plt.show()
           
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
sorted_data = df["ConvertedComp"].sort_values()
q1, q3 = np.percentile(sorted_data, [25, 75])
iqr = q3 - q1
print(iqr)
```

Find out the upper and lower bounds.


```python
# your code goes here
Upper_bound = q3 + (1.5 * iqr)
Lower_bound = q1 - (1.5 * iqr)
print(Upper_bound, Lower_bound)   
```

```python
df.boxplot(column=["ConvertedComp"])
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
Found_Outliers = []
for i in zip(df["ConvertedComp"]):
    if i > Upper_bound or i < Lower_bound:
        Found_Outliers.append(i)
        
print(len(Found_Outliers))
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
df_NewConvertedComp = df[[x not in Found_Outliers for x in zip(df["ConvertedComp"])]]
df_NewConvertedComp
```
